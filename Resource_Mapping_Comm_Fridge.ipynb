```python
# Upload the local mhtml to the Colab folder first
file_name = 'Community Fridges.mhtml'
```

```python
from bs4 import BeautifulSoup
import email
import pandas as pd

# Read the MHTML file content
with open(file_name, 'rb') as file:
    mhtml_content = file.read()

# Parse the MHTML content using the email package
msg = email.message_from_bytes(mhtml_content)
```

```python
# Extract the HTML part
html_content = None
for part in msg.walk():
    if part.get_content_type() == 'text/html':
        charset = part.get_content_charset() if part.get_content_charset() else 'utf-8'
        html_content = part.get_payload(decode=True).decode(charset)
        break

# Check if HTML content was extracted
if html_content is None:
    raise ValueError("No HTML content found in the MHTML file")

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Inspect the HTML structure to identify the relevant tags and classes
print(soup.prettify())  # Print the first 2000 characters for inspection
```

```python
# Find the table rows, excluding the header row
rows = soup.find_all('tr')[1:]

data = []

rows
```

```python
for row in rows:
    cols = row.find_all('td')
    fridge_name = cols[0].get_text(strip=True).replace('\r\n', ' ')
    address = cols[1].get_text(strip=True).replace('\n', ' ')
    borough = cols[2].get_text(strip=True) if len(cols) > 2 else None
    contact = cols[3].get_text(strip=True) if len(cols) > 3 else None
    social_media = cols[4].find('a')['href'] if len(cols) > 4 and cols[4].find('a') else None
    data.append({
        'Fridge Name': fridge_name,
        'Address': address,
        'Borough': borough,
        'Contact': contact,
        'Social Media': social_media
    })

# Convert to DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
# Export the DataFrame
df.to_excel('NYC Community Fridges.xlsx', index=False)
```
